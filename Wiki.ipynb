```python
from transformers import AutoTokenizer
from datasets import load_dataset  # Import load_dataset here

# Load the French and English Wikipedia datasets
french_dataset = load_dataset("wikimedia/wikipedia", "20231101.fr")
english_dataset = load_dataset("wikimedia/wikipedia", "20231101.en")

tokenizer = AutoTokenizer.from_pretrained("bert-base-multilingual-cased")

def preprocess_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)

french_dataset = french_dataset.map(preprocess_function, batched=True)
english_dataset = english_dataset.map(preprocess_function, batched=True)
```
